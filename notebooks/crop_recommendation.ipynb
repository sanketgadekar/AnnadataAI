```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("../data/recommendation/raw/Crop_recommendation.csv")
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report , accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('label', axis=1).reset_index(drop=True)
y = df['label'].reset_index(drop=True)
```

```python
X_train ,X_test,y_train,y_test = train_test_split(
    X,y,test_size = 0.2,random_state = 5 ,stratify = y
)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=200,max_depth=None,min_samples_split=2, min_samples_leaf=1, random_state=42)

scores = cross_val_score(model, X_train, y_train, cv=5)  # 5-fold CV

model.fit(X_train,y_train)
print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
```

```python
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, X, y, cv=cv)
model.fit(X_train,y_train)
print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
```

```python
y_pred = model.predict(X_test)

# Accuracy
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc*100:.2f}%")

# Detailed metrics
#print("\nClassification Report:\n", classification_report(y_test, y_pred))

train_acc = model.score(X_train, y_train)
test_acc = model.score(X_test, y_test)

print("Train Accuracy:", train_acc)
print("Test Accuracy:", test_acc)
```
